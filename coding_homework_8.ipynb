```python
# import modules

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF
from pandas import DataFrame as df
```

```python
# load data

fear_traces = np.loadtxt('https://raw.githubusercontent.com/avaowitt/NEURO/main/fear_conditioning_traces_hw8.csv',delimiter = ',')
extinction_predictors = np.loadtxt('https://raw.githubusercontent.com/avaowitt/NEURO/main/extinction_predictors_hw8.csv',delimiter = ',')
shock_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/shock_index.csv', delimiter = ',').astype(int)
tone_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/tone_index.csv', delimiter = ',').astype(int)
```

```python
# make two arrays from fear traces

conditioning_traces = fear_traces[:10458,:]

extinction_traces = fear_traces[10459:,:]
```

```python
# shape of arrays

print("Conditioning Traces Size:", conditioning_traces.size)
print("Extinction Traces Size:", extinction_traces.size)
print("Conditioning Traces Shape:", conditioning_traces.shape)
print("Extinction Traces Shape:", extinction_traces.shape)
```

```python
# creating time variables

time_conditioning = np.linspace(0,697.2,10458)
print(10458/15) # length of conditioning session
time_extinction = np.linspace(0,947.2666666666667,14209)
print(14209/15) # length of extinction session
```

```python
#In a new Code Cell, plot the first and second ROIs from your conditioning_traces variable in one subplot. Plot the first and second ROIs from your extinction_traces variable in a second subplot. Make the first ROI blue, and the second ROI red in each subplot. Put time on the x-axis of both subplots. Make vertical black lines at shock onset (3 total), and vertical green lines at tone onset (3 total) in your first subplot. Label your axes.

fig, (ax1, ax2)= plt.subplots(2)

ax1.plot(time_conditioning, conditioning_traces[:,0], color = 'blue')
ax1.plot(time_conditioning, conditioning_traces[:,1], color = 'red' )
ax1.axvline(np.all(shock_index),color='black')
ax1.axvline(np.all(tone_index),color='green')
ax1.set_xlabel('Time (s)')
ax1.set_yticks([])
ax1.legend({'ROI1', 'ROI2'})


ax2.plot(time_extinction, extinction_traces[:,0], color = 'blue')
ax2.plot(time_extinction, extinction_traces[:,1], color = 'red')

ax2.set_xlabel('Time (s)')
ax2.set_yticks([])
ax2.legend({'ROI1', 'ROI2'})
```

Describe the relationship between the conditioning/extinction traces for your first ROI, and the conditioning/extinction traces for your second ROI. Does there seem to be a relationship between how a cell responds during conditioning, and how it responds during extinction?

```python
# make a predictors conditioning variable

shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
predictors_conditioning['shock onset'] = shock_array
```

```python
# create r_model_conditioning variable and run for loop

r_model_conditioning = np.zeros(63)
result_list = []

for i in range(63):
  trace_temp = conditioning_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r3 = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = r3.statistic
  result_list.append(r_model_conditioning)

r_index_conditioning = np.array(result_list)
```

```python
# create r_model_extinction variable and run for loop

shock_array1 = np.zeros(14209)
for i in range(14209):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array1[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array1[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array1[i] = 1
  else:
    shock_array1[i] = 0

tone_array1 = np.zeros(14209)
for i in range(14209):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array1[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array1[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array1[i] = 1
  else:
    tone_array1[i] = 0

# make a predictors extinction variable

predictors_extinction = df(data={'Intercept': np.ones_like(tone_array1), 'tone onset': tone_array1})
predictors_extinction['shock onset'] = shock_array1


r_model_extinction = np.zeros(63)
result_list1 = []

for i in range(63):
  trace_temp1 = extinction_traces[:,i]
  model31 = sm.GLM(trace_temp1, predictors_extinction, family=Poisson())
  model31_results = model31.fit()
  predicted_trace1 = model31_results.predict()
  r31 = pearsonr(trace_temp1, predicted_trace1)
  r_model_extinction[i] = r31.statistic
  result_list1.append(r_model_extinction)

r_index_extinction = np.array(result_list1)
```

```python
# correlation between r_index_conditioning and r_index_extinction?

correlation_coefficient, p_value = pearsonr(r_index_conditioning, r_index_extinction)
print(correlation_coefficient)
print(p_value)
```

```python
# scatter plot

plt.scatter(r_index_conditioning, r_index_extinction)
```

Interpret your results. Based on your correlation coefficient and scatter plot, does it seem like a cell that responds strongly to shock and tone onset during the conditioning session is more likely to respond strongly to tone onset during the extinction session?

More likely to respond to the shock and tone during the conditioning session. This is because the correlation is higher (generally) for responding to conditioning.
