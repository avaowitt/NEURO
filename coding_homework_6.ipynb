```python
# loading modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data

accfa = 'https://raw.githubusercontent.com/avaowitt/NEURO/main/acc_fa_hw6.csv'
acc_fa = np.loadtxt(accfa, delimiter = ',')

acchit = 'https://raw.githubusercontent.com/avaowitt/NEURO/main/acc_hits_hw6.csv'
acc_hit = np.loadtxt(acchit, delimiter = ',')

lcfa = 'https://raw.githubusercontent.com/avaowitt/NEURO/main/lc_fa_hw6.csv'
lc_fa = np.loadtxt(lcfa, delimiter = ',')

lchit = 'https://raw.githubusercontent.com/avaowitt/NEURO/main/lc_hits_hw6.csv'
lc_hit = np.loadtxt(lchit, delimiter = ',')
```

```python
# shape of arrays

print(acc_fa.shape)
print(acc_hit.shape)
print(lc_fa.shape)
print(lc_hit.shape)
```

```python
# time variable

time = np.linspace(0,4,8001)
```

```python
# finding sampling rate

lengthofsample = 4
numberofsamples = 8001

samplingrate = numberofsamples/lengthofsample
print(samplingrate)
```

```python
# graphing!

fig, (ax1, ax2)= plt.subplots(1,2)

ax1.plot(time, acc_hit[0], color='red')
ax1.plot(time, lc_hit[0], color='blue')
ax1.set_xlabel('Time')
ax1.set_ylabel('Hits')
ax2.plot(time, acc_fa[0], color='red')
ax2.plot(time, lc_fa[0], color='blue')
ax2.set_xlabel('Time')
ax2.set_ylabel('Failed Hits')
plt.show()
```

Recall from our notebook on phase coherence that sometimes the peaks and troughs of two oscillations have a consistent relationship across time. Is there any evidence of this in the raw LFP?

```python
# compute coherence for HITS

dt = time[1] - time[0]
N = acc_hit.shape[1]

# Compute the Fourier transforms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hit])  # ... for each trial in E1
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hit])  # ... and each trial in E2

# Compute the spectra
T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of E1 trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and E2 trials
Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     # Define a frequency axis.
plt.plot(f, cohr.real)                      # Plot coherence vs frequency,
plt.xlim([0, 50])                           # ... in a chosen frequency range,
plt.ylim([0, 1])                            # ... with y-axis scaled,
plt.xlabel('Frequency [Hz]')                # ... and with axes labeled.
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
```

Describe your results. In which frequency band is there the most phase coherence during hits?

Between 1-2Hz and 6-8Hz is the highest coherence. There is varying phase coherence from 10-50Hz while on a slope downward. It should be noted that coherence drops significantly between 2-6Hz, reaching about the average coherence.

```python
# compute coherence for FAILURES
N2 = acc_fa.shape[1]

# Compute the Fourier transforms
xf2 = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])  # ... for each trial in E1
yf2 = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])  # ... and each trial in E2

# Compute the spectra
T = time[-1]
scale = 2 * dt**2 / T
Sxx2 = scale * (xf2 * xf2.conj()).mean(0)  # Spectrum of E1 trials
Syy2 = scale * (yf2 * yf2.conj()).mean(0)  # ... and E2 trials
Sxy2 = scale * (xf2 * yf2.conj()).mean(0)  # ... and the cross spectrum

# Compute the coherence.
cohr2 = np.abs(Sxy2) / (np.sqrt(Sxx2) * np.sqrt(Syy2))

f2 = np.fft.rfftfreq(N2, dt)                     # Define a frequency axis.
plt.plot(f2, cohr2.real)                      # Plot coherence vs frequency,
plt.xlim([0, 50])                           # ... in a chosen frequency range,
plt.ylim([0, 1])                            # ... with y-axis scaled,
plt.xlabel('Frequency [Hz]')                # ... and with axes labeled.
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
```

```python
# coherence on graph for COMPARISON

plt.plot(f, cohr.real, color='red')                     # Plot coherence vs frequency,
plt.plot(f2,cohr2.real, color='blue')
plt.xlim([0, 50])                           # ... in a chosen frequency range,
plt.ylim([0, 1])                            # ... with y-axis scaled,
plt.xlabel('Frequency [Hz]')                # ... and with axes labeled.
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
```
