```python
# loading modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data

spl1 = 'https://raw.githubusercontent.com/avaowitt/NEURO/main/spikes_left_trial1.csv'
spikes_left_trial1 = np.loadtxt(spl1, delimiter = ',')

spl2 = 'https://raw.githubusercontent.com/avaowitt/NEURO/main/spikes_left_trial2.csv'
spikes_left_trial2 = np.loadtxt(spl2, delimiter = ',')

spr1 = 'https://raw.githubusercontent.com/avaowitt/NEURO/main/spikes_right_trial1.csv'
spikes_right_trial1 = np.loadtxt(spr1, delimiter = ',')

spr2 = 'https://raw.githubusercontent.com/avaowitt/NEURO/main/spikes_right_trial2.csv'
spikes_right_trial2 = np.loadtxt(spr2, delimiter = ',')
```

```python
# shape of arrays

print(len(spikes_left_trial1))
print(len(spikes_left_trial2))
print(len(spikes_right_trial1))
print(len(spikes_right_trial2))
```

```python
# determine the firing rate

fireratel1 = (len(spikes_left_trial1)/30)
print(fireratel1)
fireratel2 = (len(spikes_left_trial2)/30)
print(fireratel2)
firerater1 = (len(spikes_right_trial1)/30)
print(firerater1)
firerater2 = (len(spikes_right_trial2)/30)
print(firerater2)
```

Interpret these results - do you think this neuron fires more before left turn trials, or right turn trials?

This is an interesting result because the data reveals that the neuron fired more before the left turn on trial one, but more before the right turn on trial 2.

```python
# plot the data!

fig, (ax1, ax2)= plt.subplots(1,2)

ax1.plot(spikes_left_trial1, np.ones_like(spikes_left_trial1), '.')
ax1.plot(spikes_right_trial1, 2 * np.ones_like(spikes_right_trial1), '.')
ax1.set_xlim([0, 5])
ax1.set_xlabel('Time (s)')
ax1.set_yticks([])
ax1.legend({'Left', 'Right'})


ax2.plot(spikes_left_trial2, np.ones_like(spikes_left_trial2), '.')
ax2.plot(spikes_right_trial2, 2 * np.ones_like(spikes_right_trial2), '.')
ax2.set_xlim([0, 5])
ax2.set_xlabel('Time (s)')
ax2.set_yticks([])
ax2.legend({'Left', 'Right'})
```

Interpret the graph from step 8. Does this neuron seem more "bursty", or more regular-firing, during left or right turn trials?

This neuron fires much more (bursty) during left turn trials. It also appears that the second trial resulted in greater firing of this neuron during both left and right trials. This makes sense as the neuron is gaining strength through the knowledge gained from trial one. It also appears that the right turn trials fill in the quote unquote gaps or intervals where this neuron is not firing, which again makes sense considering the neuron is being used to complete a different task (left vs right).

```python
# creating ISI arrays

spikes_left_trial1ISI = np.diff(spikes_left_trial1)
spikes_left_trial2ISI = np.diff(spikes_left_trial2)
spikes_right_trial1ISI = np.diff(spikes_right_trial1)
spikes_right_trial2ISI = np.diff(spikes_right_trial2)
```

```python
# histograms of the ISI's - with full xlim according to bins

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

bins = np.arange(0, 2, 0.01)  # Define the bins for the histogram

ax1.hist(spikes_left_trial1ISI, bins)        # Plot the histogram of the ISI data
ax1.set_xlim([0, 2])            # ... focus on ISIs from 0 to 150 ms
ax1.set_xlabel('ISI [s]')              # ... label the x-axis
ax1.set_ylabel('Counts')               # ... and the y-axis
ax1.set_title('Left1')

ax2.hist(spikes_left_trial2ISI, bins)        # Plot the histogram of the ISI data
ax2.set_xlim([0, 2])            # ... focus on ISIs from 0 to 150 ms
ax2.set_xlabel('ISI [s]')              # ... label the x-axis
ax2.set_ylabel('Counts')               # ... and the y-axis
ax2.set_title('Left2')

ax3.hist(spikes_right_trial1ISI, bins)        # Plot the histogram of the ISI data
ax3.set_xlim([0, 2])            # ... focus on ISIs from 0 to 150 ms
ax3.set_xlabel('ISI [s]')              # ... label the x-axis
ax3.set_ylabel('Counts')               # ... and the y-axis
ax3.set_title('Right1')

ax4.hist(spikes_right_trial2ISI, bins)        # Plot the histogram of the ISI data
ax4.set_xlim([0, 2])            # ... focus on ISIs from 0 to 150 ms
ax4.set_xlabel('ISI [s]')              # ... label the x-axis
ax4.set_ylabel('Counts')               # ... and the y-axis
ax4.set_title('Right2')
```

```python
# histograms of the ISI's - narrowed down to the first 0.5 seconds since that is where most of the bursts are occuring

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

bins = np.arange(0, 2, 0.01)

ax1.hist(spikes_left_trial1ISI, bins)
ax1.set_xlim([0, 0.5])
ax1.set_xlabel('ISI [s]')
ax1.set_ylabel('Counts')
ax1.set_title('Left1')

ax2.hist(spikes_left_trial2ISI, bins)
ax2.set_xlim([0, 0.5])
ax2.set_xlabel('ISI [s]')
ax2.set_ylabel('Counts')
ax2.set_title('Left2')

ax3.hist(spikes_right_trial1ISI, bins)
ax3.set_xlim([0, 0.5])
ax3.set_xlabel('ISI [s]')
ax3.set_ylabel('Counts')
ax3.set_title('Right1')

ax4.hist(spikes_right_trial2ISI, bins)
ax4.set_xlim([0, 0.5])
ax4.set_xlabel('ISI [s]')
ax4.set_ylabel('Counts')
ax4.set_title('Right2')
```

Interpret your histograms. Does this neuron tend to fire in "bursts"? Does it have higher "burstiness" during left or right turn trials?

Yes this neuron does tend to fire in bursts. The first left trial appears more bursty than the first right trial. This is consistent throughout the second trial as well. The right trial is also bursty but bursts more consistently over a longer period of time than the left trials which have high bursts concentrated between 0-0.2 seconds. Both histograms tend to fill out (get more bursty) during the second trials.

```python
# calculate firing rate and graph it
binsize = 1                                                                 # Let's start with 1-second bins
fr_bins = np.arange(0,30,binsize)                                             # Make time bins (in 1 second increments) from 0 to 30 s
fr_l = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index = (spikes_left_trial1ISI > fr_bins[i]) & (spikes_left_trial1ISI < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_left_trial1ISI[spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_l[i] = (len(true_values))/binsize

fr_l2 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index = (spikes_left_trial2ISI > fr_bins[i]) & (spikes_left_trial2ISI < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_left_trial2ISI[spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_l2[i] = (len(true_values))/binsize

fr_r = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index = (spikes_right_trial1ISI > fr_bins[i]) & (spikes_right_trial1ISI < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_right_trial1ISI[spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_r[i] = (len(true_values))/binsize

fr_r2 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index = (spikes_right_trial2ISI > fr_bins[i]) & (spikes_right_trial2ISI < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_right_trial2ISI[spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_r2[i] = (len(true_values))/binsize

fig, (ax1, ax2)= plt.subplots(1,2)

ax1.plot(fr_bins[:-1], fr_l, color='red')
ax1.plot(fr_bins[:-1], fr_r, color='blue')
ax1.set_xlim([0, 30])
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Firing Rate [Hz]')
ax1.set_yticks([])
ax1.legend({'Left', 'Right'})


ax2.plot(fr_bins[:-1], fr_l2, color='red')
ax2.plot(fr_bins[:-1], fr_r2, color='blue')
ax2.set_xlim([0, 30])
ax2.set_xlabel('Time (s)')
ax2.set_yticks([])
ax2.legend({'Left', 'Right'})
```

```python
# same previous code cell - graphs refined to a 10 second span to see the graph differences more clearly

fig, (ax1, ax2)= plt.subplots(1,2)

ax1.plot(fr_bins[:-1], fr_l, color='red')
ax1.plot(fr_bins[:-1], fr_r, color='blue')
ax1.set_xlim([0, 10])
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Firing Rate [Hz]')
ax1.set_yticks([])
ax1.legend({'Left', 'Right'})


ax2.plot(fr_bins[:-1], fr_l2, color='red')
ax2.plot(fr_bins[:-1], fr_r2, color='blue')
ax2.set_xlim([0, 10])
ax2.set_xlabel('Time (s)')
ax2.set_yticks([])
ax2.legend({'Left', 'Right'})
```

During trial two the firing rate is pretty much the same for both left and right turns. During trial one they are a little more differentiated, such as being opposites of each other at four seconds and eight seconds, yet this is smoothed out during the second trial.
