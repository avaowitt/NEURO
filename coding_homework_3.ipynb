```python
# import programs

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# import neuro data as numpy arrays!

v_rest_sst = np.array(np.loadtxt('https://raw.githubusercontent.com/avaowitt/NEURO/main/v_rest_sst.csv'))

v_rest_pvalb = np.array(np.loadtxt('https://raw.githubusercontent.com/avaowitt/NEURO/main/v_rest_pvalb.csv'))
```

```python
# find means

mean_sst = np.mean(v_rest_sst)
print(mean_sst)
mean_pvalb = np.mean(v_rest_pvalb)
print(mean_pvalb)

# t-test of null hypothesis

print(stats.ttest_ind(v_rest_sst,v_rest_pvalb))

# plot graphs IF p-value < 0.05 and label axes

statistic, p_value = stats.ttest_ind(v_rest_sst,v_rest_pvalb)
print(p_value)

if p_value <.05:
  fig, (ax1, ax2)= plt.subplots(1,2)

  ax1.hist(v_rest_sst)
  ax1.set_ylabel('Frequency')
  ax1.set_xlabel('Resting Membrane Potential SST')
  ax2.hist(v_rest_pvalb)
  ax2.set_xlabel('Resting Membrane Potential PVALB')
  plt.show()

else:
  print('no graph womp womp')
```
